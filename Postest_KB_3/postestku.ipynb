```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('supermarket_sales - Sheet1.csv')
data
```

```python
data.keys()
```

# 1.  GRAFIK BATANG

```python
a = data.groupby("Product line")[['Product line','Quantity']].sum()
a.reset_index(inplace=True)
```

```python
plt.bar(a['Product line'], a['Quantity'], color="#8B0000")
plt.xlabel("Product line")
plt.ylabel('Quantity')
plt.title("Jumlah Data Berdasarkan  Kategori Produk")
plt.xticks(rotation = 80)
plt.show()
```

# Dari grafik di atas terlihat perbandingan jumlah data setiap kategori produk , yang dimana rata-rata jumlah produk berada di angka 800-1000.Dari grafik di atas juga dapat disimpulkan bahwa produk kategori elektronik memiliki jumlah tertinggi dibandingkan kategori produk lain.

```python
data["City"].value_counts()
```

# 2. GRAFIK PIE

```python
data_pie = data["City"].value_counts().rename_axis("City").reset_index(name="jumlah")
```

```python
data_pie.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie["jumlah"],labels=data_pie['City'],startangle=90,autopct='%.1f%%',shadow=True ,explode=(0.05,0,0))

plt.title('persentase lokasi supermarket berdasarkan di setiap kota')
plt.legend(data.index,title='Product line')
plt.show()
```

# KESIMPULAN
# Dari  grafik diatas dapat disimpulkan persentase perbandingan jumlah cabang supermarket berdasarkan lokasi ("kota").yang dimana cabang supermarket terbanyak berada di kota Yangon 


# 3. GRAFIK GARIS

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
ind = (pd.DataFrame(data.groupby('Date')['Total'].sum()))
ind.reset_index('Date')
ind.head()
```

```python
ind.resample('M')
```

```python
plt.figure(figsize=(15,10))
plt.plot(ind.index, ind)
plt.legend(['Date', 'Total']) #untuk menampilkan keterangan
plt.title("Perkembangan Total kinerja penjulalan dari waktu ke waktu")
plt.xlabel('Date')
plt.ylabel('Total')
plt.show()
```

# KESIMPULAN
# Dari graffik di atas terlihat persentase total pendapatan dari waktu ke waktu yang dimana pendapatan tertinggi  dari grafik diatas berada di kurun waktu 15-03-2019 dan pendapatan terendah berada di kurun waktu 15-02-2019
