# NORMALISASI

```python
dt_normalisasi = data.copy()
dt_normalisasi['odometer_value'] = (dt_normalisasi['odometer_value'] - dt_normalisasi['odometer_value'].min()) / (dt_normalisasi['odometer_value'].max() - dt_normalisasi['odometer_value'].min())
dt_normalisasi['year_produced']   = (dt_normalisasi['year_produced'] - dt_normalisasi['year_produced'].min()) / (dt_normalisasi['year_produced'].max() - dt_normalisasi['year_produced'].min())
print(dt_normalisasi['odometer_value'])
print(dt_normalisasi['year_produced'])
```
